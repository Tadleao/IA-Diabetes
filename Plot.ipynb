```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import seaborn as sns
```

```python
# Assuming 'data' is your n-dimensional dataset
# data.shape should be (n_samples, n_features)

# Reduce the dimensionality to 2 using PCA
df = pd.read_csv('balanced_over.csv')
df_std = (df - df.mean()) / df.std()
df_std
```

```python
pca = PCA(n_components=2)
pca_result = pca.fit_transform(df_std)
pca_df = pd.DataFrame(data=pca_result, columns=['PC1', 'PC2'])
plt.figure(figsize=(8, 6))
plt.scatter(pca_df['PC1'], pca_df['PC2'], c=df['Q03001'].map({1: '#ba0909', 2: '#3452bf'}))
plt.title('2D Plot of N-dimensional DataFrame')
plt.grid(True)
plt.show()
# Plot the data points in 2D
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(pca_df['PC1'], pca_df['PC2'])
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('2D Plot of N-dimensional DataFrame')
plt.grid(True)
plt.show()
```
