```python
import pandas as pd
from scipy import stats
from sklearn.impute import KNNImputer
from sklearn.preprocessing import MinMaxScaler
```

```python
#Ler Base
df = pd.read_csv("pns2019.csv")
#Array de atributos selecionados
attr =['C006','C008','I00102','Q046017','Q046015','Q046018','Q03802','Q02901','Q03002','Q03001','N001','N00101','N011','P00103','P00404','P02401','P02002','P018','P023','P027','P035','P05401','P052','P053','P050','P058','VDF004']
i = 0
df = df[attr]
df = df.drop_duplicates()
df
```

```python
#Base filtrada para csv
df.to_csv('filtrada.csv')
```

```python
# Base apenas com quem respondeu que recebeu diagnóstico com Diabetes
mask = df['Q03001'] == 1.0
diab = df[mask]
# Base apenas com quem respondeu que não recebeu diagnóstico com Diabetes
mask2 = df['Q03001'] == 2.0
ndiab = df[mask2]
```

```python
# Passando ambas para o csv
diab.to_csv('DIABETES.csv')
ndiab.to_csv('N_DIABETES.csv')
```

```python
full=pd.concat([diab,ndiab], ignore_index=True, axis=0)
full
```

```python
full.to_csv('Full_Diab_NDiab.csv', index=False)
```

```python
imputer = KNNImputer(n_neighbors=2)
full_input = imputer.fit_transform(full)
full_input
```

```python
full_input = pd.DataFrame(full_input)
full_input.to_csv('inputed.csv', index=False)
full_input
```

```python
full_input = pd.read_csv('inputed.csv', index_col=False)
z_scores = stats.zscore(full_input, nan_policy='omit')

# Define threshold (e.g., |Z| > 3)
threshold = 3

# Remove rows with outliers
df_no_outliers = full_input[(z_scores < threshold).all(axis=1)]
columns_mapping = dict(zip(df_no_outliers.columns, df.columns))
df_no_outliers.reset_index(drop=True, inplace=True)
df_no_outliers.rename(columns=columns_mapping, inplace=True)
df_no_outliers.to_csv('noOut.csv', index=False)
df_no_outliers
```

```python
scaler = MinMaxScaler()
cl = ['C008', 'P00103', 'P00404', 'P053']
df_no_outliers[cl] = pd.DataFrame(scaler.fit_transform(df_no_outliers[cl]))
df_no_outliers
```

```python
df_no_outliers.to_csv('normalized.csv', index=False)
```
