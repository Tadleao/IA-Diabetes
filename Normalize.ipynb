```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('noOut.csv')
df
```

```python
scaler = MinMaxScaler()
cl = ['C008', 'P00103', 'P00404', 'P053']
df[cl] = pd.DataFrame(scaler.fit_transform(df[cl]))
df
```

```python
df.to_csv('normalized.csv', index=False)
```
