```python
import pandas as pd
from imblearn.under_sampling import NearMiss
```

```python
df = pd.read_csv('normalized.csv')
df
```

```python
X = df[df.columns]
X = X.drop('Q03001', axis=1)
y = df['Q03001']
```

```python
nm = NearMiss()
X_resampled, y_resampled = nm.fit_resample(X, y)
df_resampled = pd.concat([pd.DataFrame(X_resampled, columns=['C006', 'C008', 'I00102', 'J01101', 'N001', 'N00101', 'N011', 'P00103',
       'P00404', 'P00610', 'P00611', 'P018', 'P02002', 'P023', 'P02401',
       'P027', 'P035', 'P050', 'P052', 'P053', 'P05401', 'P058', 'Q02901',
       'Q03002', 'Q03802', 'Q046015', 'Q046017', 'Q046018',
       'VDF004']), pd.DataFrame(y_resampled, columns=['Q03001'])], axis=1)
df_resampled
```

```python
df_resampled.to_csv("balanced_under.csv", index=False)
```
